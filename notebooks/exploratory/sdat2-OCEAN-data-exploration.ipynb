```python
import os
import xarray as xr
from src.constants import OCEAN_DATA_PATH, OCEAN_PATH, OCEAN_OUTPUT_PATH
import matplotlib.pyplot as plt
```

### Plot surface / first time step for each netcdf

```python
for item in [x for x in os.listdir(OCEAN_DATA_PATH) if ".nc" in x]:
    name = OCEAN_DATA_PATH / item
    xr_da = xr.open_dataarray(name, decode_times=False)
    print(xr_da)
    if "T" in xr_da.dims:
        xr_da = xr_da.isel(T=0)
    if "Z" in xr_da.dims:
        xr_da = xr_da.isel(Z=0)
    print(name)
    print(xr_da)
    xr_da.plot()
    plt.show()
    print(name)
    plt.clf()
```

```python
OCEAN_OUTPUT_PATH = OCEAN_PATH / "RUN" / "output"
```

```python
for item in [x for x in os.listdir(OCEAN_OUTPUT_PATH) if ".nc" in x]:
    name = OCEAN_OUTPUT_PATH / item
    print(name)
    xr_ds = xr.open_dataset(name, decode_times=False)
    print(xr_ds)
    for var in [x for x in xr_ds.variables if "0" not in x and "GRID" not in x]:
        xr_da = xr_ds[var]
        print(xr_da)
        T_coord = [x for x in xr_da.dims if "T" in x][0]
        L_coord = [x for x in xr_da.dims if "L" in x][0]
        print(T_coord, L_coord)
        xr_da.isel(indexers={L_coord: 0, T_coord: 0}).plot()
        plt.show()
        plt.clf()
```
