```python
from src.utils import open_dataarray
```

```python
qflx = open_dataarray(os.path.join(OCEAN_DATA_PATH, "qflx.nc"))
qflx_0 = open_dataarray(os.path.join(OCEAN_DATA_PATH, "qflx-0.nc"))
diff = qflx - qflx_0
```

```python
import matplotlib.pyplot as plt
```

```python
for t in range(12):
    vmin, vmax = -0.0002, 0.0002
    fix, axs = plt.subplots(3, 1, figsize=[3, 7])
    qflx.isel(T=t).plot(ax=axs[0], vmin=vmin, vmax=vmax, cmap="RdBu")
    qflx_0.isel(T=t).plot(ax=axs[1], vmin=vmin, vmax=vmax, cmap="RdBu")
    diff.isel(T=t).plot(ax=axs[2], vmin=vmin, vmax=vmax, cmap="RdBu")
    plt.tight_layout()
    plt.show()
```
