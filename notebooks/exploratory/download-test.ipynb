```python
from src.utils import timeit
import zipfile
from tqdm import tqdm
```

```python
lol = [
    [
        "atmos",
        [
            ["https://www.dropbox.com/s/j7x3bjfnb8fdw3b/tmp.zip?raw=1", "tmp.zip"],
            ["https://www.dropbox.com/s/r82fv6g8sdwykzj/DATA.zip?raw=1", "DATA.zip"],
        ],
    ],
    [
        "ocean",
        [
            ["https://www.dropbox.com/s/luoo402lzbsyi5r/DATA.zip?raw=1", "DATA.zip"],
            [
                "https://www.dropbox.com/s/q2qxeucdsb9958j/output.zip?raw=1",
                "output.zip",
            ],
        ],
    ],
]
```

```python
for item in lol:
    direc = item[0]
    if not os.path.exists(direc):
        os.mkdir(direc)
    for pair in item[1]:
        url = pair[0]
        name = pair[1]
        get_unzip(direc, url, name)
```

```python
@timeit
def get_unzip(direc: str, url: str, name: str) -> None:
    @timeit
    def get_zip() -> None:
        write_path = os.path.join(direc, name)
        req = requests.get(url, stream=True)
        with open(write_path, "wb") as file:
            for chunk in tqdm(req.iter_content(chunk_size=128)):
                file.write(chunk)

    @timeit
    def un_zip() -> None:
        write_path = os.path.join(direc, name)
        with zipfile.ZipFile(write_path, "r") as zip_ref:
            zip_ref.extractall(direc)

    get_zip()
    un_zip()
```
